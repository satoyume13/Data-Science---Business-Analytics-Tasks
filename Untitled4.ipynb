### **Prediction using Supervised ML**

This is a Simple ML Project to Predict the percentage of an student based on the no. of study hours.

```python
# Importing all libraries required in this notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

# Read and View Data

```python
data = pd.read_csv('http://bit.ly/w-data')
data.head(20)
```

```python
#View data details
data.describe()
```

# Plotting the distribution of scores

```python
data.plot(x='Hours', y='Scores', style='o', color='green')
plt.title('Hours vs Percentage', size = 20, color = 'red')
plt.xlabel('Hours Studied', size =13, color = 'blue')
plt.ylabel('Percentage Score', size = 13, color = 'blue')


plt.grid()
plt.show()
```

```python
#Preparing the data
X = data.iloc[:, :-1].values
y = data.iloc[:, 1].values

#split the data
train_x, test_x, train_y, test_y = train_test_split(X, y,
                            test_size=0.2, random_state=0)
```

# Training the Algorithm

```python
regr = LinearRegression()
regr.fit(train_x, train_y)


print("Model Trained Succssefuly.")
```

# Plotting the regression line

```python
# Plotting the regression line
line = regr.coef_*X+regr.intercept_

# Plotting the test data
plt.scatter(X, y, color = 'red')
plt.plot(X, line);
plt.grid()
plt.show()
```

# Making Predictions

```python
#Testing hours data
print(test_x)
#Score predicton
y_pred = regr.predict(test_x)
```

# Comparing Actual vs Predicted

```python

pridict = pd.DataFrame({'Hours': test_y, 'Predicted': y_pred})
pridict
```

# Testing provided Data

```python
hours = 9.25
own_pred = regr.predict([[hours]])
print("No of Hours =", hours, ",Predicted Score =", own_pred[0])
```

#  Model Evaluation

```python
er=mean_absolute_error(test_y,y_pred)
print('Mean Absolute Error=',er)
```
